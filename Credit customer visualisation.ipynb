```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
object_features = df.select_dtypes(['object']).columns.to_list()
for obj__features in object_features:
    fig, ax = plt.subplots(figsize=(6, 6), facecolor = 'white')
    plt.title(obj__features, fontsize=25, fontweight='bold', color= 'black')
    eda_percentage = df.groupby(obj__features).size()
    sns.set()
    eda_percentage.plot(kind='pie', autopct='%.0f%%')
    plt.ylabel("")
    plt.show()
```

```python
continous_features = df.select_dtypes(['float64']).columns.to_list()
for cont_feature in continous_features: 
    plt.figure()
    plt.title(cont_feature)
    ax = sns.histplot(df[cont_feature],kde=True, stat="density", linewidth=0 )
    plt.show()
```

```python
discrete_features = df.select_dtypes(['int64']).columns.to_list()
for disc_feature in discrete_features: 
    plt.figure()
    plt.title(disc_feature)
    ax = sns.histplot(df[disc_feature],kde = False)
```
